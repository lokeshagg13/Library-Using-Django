```python
from selenium import webdriver
import requests
import re
import pandas as pd
import time
import json

from random import randrange
from datetime import timedelta
from datetime import datetime
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('http://www.topshelfcomix.com/catalog/isbn-list')
```

```python
isbns = [x.text for x in driver.find_elements_by_class_name('isbn-number')]
```

```python
len(isbns)
```

```python
isbns = isbns[1:]
```

```python
for i in range(len(isbns)):
    isbns[i] = re.sub('[^0-9]','', isbns[i])
```

```python
newisbns = []
for i in range(len(isbns)):
    if(len(isbns[i])==10 or len(isbns[i])==13):
        newisbns.append(isbns[i])
        
```

```python
len(newisbns)
```

```python
isbns
```

```python
df = pd.DataFrame(columns = ['Sno','Title','Summary','ISBN','Author','language','genre'])
languages = ['English','French','Russian','Mandarin','Spanish','Hindi']
```

```python
for i in range(len(isbns)):
    time.sleep(2)
    resp = requests.get('http://openlibrary.org/api/books?bibkeys=ISBN:'+isbns[i]+'&jscmd=data&format=json')
    if resp.status_code==200:
        resp = resp.json()
        print(resp.keys())
        if(len(list(resp.keys()))>0):
            resp = resp[list(resp.keys())[0]]
        
            if('title' in resp.keys()):
                print(resp['title'])
                df.loc[i,'Title'] = resp['title']
                df.loc[i,'Summary'] = resp['title']
            if('authors' in resp.keys()):
                print(resp['authors'][0]['name'])
                df.loc[i,'Author'] = resp['authors'][0]['name']
            if('subjects' in resp.keys()):
                genres = []
                gn = resp['subjects']
                for j in range(len(gn)):
                    for x in gn[j]['name'].split(', '):
                        genres.append(x) 
                print(genres)
                df.loc[i,'genre'] = genres

            df.loc[i,'ISBN'] = isbns[i]
            df.loc[i,'language'] = languages[i//40]
        
```

```python
df.to_csv('Book.csv',index=False)
```

```python
## To be run in command prompt after running python manage.py shell
def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)
```

```python
for i in range(len(authors)):
    authorname = authors[i].split(' ')
    dfauthors.loc[i,'First Name'] = authorname[0]
    dfauthors.loc[i,'Last Name'] = authorname[1]
```
